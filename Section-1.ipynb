2. Handling missing values

```python
employee_data.isnull().sum()
```

```python
employee_data['Name'].fillna("ABC",inplace = True)
```

```python
employee_data.isnull().sum()
```

```python
employee_data['Department'].unique()
```

```python
if 'Department' in employee_data.columns:
    mode = employee_data['Department'].mode()[0]

    employee_data['Department'].fillna(mode, inplace=True)
    
```

```python
employee_data.isnull().sum()
```

```python
employee_data['Salary'].nunique()
```

```python
if 'Salary' in employee_data.columns:
    sum = employee_data['Salary'].sum()
    avg = sum / employee_data['Salary'].nunique()
    employee_data['Salary'].fillna(avg, inplace=True)
```

```python
employee_data.head()
```

```python
employee_data.isnull().sum()
```

```python
employee_data['Status'].unique()
```

```python
if 'Status' in employee_data.columns:
    mode = employee_data['Status'].mode()[0]

    employee_data['Status'].fillna(mode, inplace=True)
```

```python
employee_data.isnull().sum()
```

3. Converting JoiningDate to proper format

```python
if 'JoiningDate' in employee_data.columns:
    employee_data['JoiningDate'] = pd.to_datetime(
        employee_data['JoiningDate'],
        infer_datetime_format=True  
    )
```

```python
employee_data
```

4. Filter out employees who are no longer working (where Status is 'Resigned')

```python
if 'Status' in employee_data.columns and 'EmployeeID' in employee_data.columns:
    mask = employee_data['Status'].fillna('').astype(str).str.strip().str.lower() == 'resigned'
    resigned = employee_data.loc[mask, ['EmployeeID', 'Name', 'Status', 'JoiningDate']]
    print("Resigned EmployeeIDs:", resigned['EmployeeID'].unique().tolist())
    display(resigned) 
```

```python
employee_data.shape
```

```python
employee_data.isnull().sum()
```

```python
if 'JoiningDate' in employee_data.columns:
    employee_data['JoiningDate'] = pd.to_datetime(
        employee_data['JoiningDate'], errors='coerce', infer_datetime_format=True
    )

    employee_data['YearsOfService'] = 2025 - employee_data['JoiningDate'].dt.year

    print(employee_data[['JoiningDate', 'YearsOfService']].head())
```

```python
employee_data.head()
```

2. Analysis

Find the average salary of employees in each department.

```python
if 'Department' in employee_data.columns and 'Salary' in employee_data.columns:
    # ensure numeric salary
    employee_data['Salary'] = pd.to_numeric(employee_data['Salary'], errors='coerce')
    dept_avg = (employee_data.groupby('Department')['Salary'].mean().round(2))
    print(dept_avg)
```

List all employees who joined after the year 2020.

```python
if 'JoiningDate' in employee_data.columns:
    joined_after_2020 = employee_data[employee_data['JoiningDate'].dt.year > 2020]
    print("Count:", len(joined_after_2020))
    display(joined_after_2020[['EmployeeID', 'Name', 'Department', 'JoiningDate']])
```

Task-2
Data Visualization

```python
data = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data.head()
```

```python
# PLot-1: Box plot to show attrition by age
sns.boxplot(x="Attrition",y="Age",data=data)
plt.title("Employee Distribution by Age & Attrition")
plt.show()
```

```python
#plot-2: employees by department
value_1 = data["Department"].value_counts()
sns.barplot(x=value_1.index, y=value_1.values)
plt.title("Employees by Department")
```

```python
#plot-3: employee distribution by education
value_1 = data["Education"].value_counts()
sns.barplot(x=value_1.index,y=value_1.values,order=value_1.index)
plt.title("Employees Distribution by Education")
plt.show()
```

```python
#plot-4: employee distribution by education field
value_1 = data["EducationField"].value_counts()
sns.barplot(x=value_1.index, y=value_1.values,order=value_1.index)
plt.title("Employees by Education Field")
```

```python
#Correlation heatmap
num = data.select_dtypes(include=[np.number])
corr = num.corr()

plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1, linewidths=0.5)
plt.title("Correlation Heatmap — Numerical Variables")
plt.tight_layout()
plt.show()
```
